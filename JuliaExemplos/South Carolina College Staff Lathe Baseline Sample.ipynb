```julia
using DataFrames
using CSV
```

```julia
# Reading in our data
df = CSV.read("SCaSa.csv")
```

```julia
showall(df)
```

```julia
# We need to get rid of the identification columns, these will be worthless for figuring out
#    compensation, and do nothing but give away information.
delete!(df, :First)
delete!(df, :Middle)
delete!(df, :Last)
```

```julia
# Now we need to get rid of Base salary, and other compensation so that the model can't catch on,
#     And cause leakage.
delete!(df, :Other_Compensation)
delete!(df, :Base_Salary)
```

```julia
# Now we have 3 columns, Agency, Title, and Total Compensation:
#   Our target is Total Compensation:
# First we'll train test split, then
#     We'll start with a Baseline
using Lathe: preprocess
```

```julia
# Train Test Splitting our Target
yTrain,yTest = preprocess.TrainTest(df.Total_Compensation)
```

```julia
#Using Lathe to get a baseline accuracy on a mean model:
using Lathe: models
```

```julia
#Constructing and fitting our MeanBaseline Model with our Trainy
#    Note we only need y because the baseline only uses the Y.
model = models.meanBaseline(yTrain)
```

```julia
# Now we can predict the model by putting in our xTest,
#   For this baseline model, we really do not need to get our xTest, it's not really vital
# For a prediction that just fills the mean in for each value.
y_pred = models.predict(model,xTest)
```
